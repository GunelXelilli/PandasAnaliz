```python
#Superstore dataset-i event-level melumatlar (visit,add_to_chart,checkout)
#saxlamadigi ucun funnel merhelelerini movcud stunlardan proxy olaraq yaratmisam
#Bu yanasma real user davranisi deyil,bisnez mentigine esaslanan texmini funnel analizini gosterir
```

```python
purchase=df[df['Sales']>0]['Customer ID'].unique() #satis
checkout=df[df['Profit']>0]['Customer ID'].unique() #kassa
cart=df[df['Discount']>0]['Customer ID'].unique() #kart

funnel={
    'Visited Product Page': len(df['Customer ID'].unique()), #her mesteri azi bir defe mehsullari
    'Added to Cart':len(set(cart)),
    'Checkout Started':len(set(cart) & set(checkout)),
    'Purchase Completed':len(set(cart) & set(checkout) & set(purchase))

}

#Dataframe seklinde
funnel_df=pd.DataFrame.from_dict(funnel,orient='index',columns=['Users'])

#Drop off faizi
#Drop-off=1-(Next step users/current step users)
funnel_df['Drop-off %']=1-funnel_df['Users'].shift(-1)/funnel_df['Users']
funnel_df['Drop-off %']=funnel_df['Drop-off %'].fillna(0)* 100

print(funnel_df)
```

# Cohort analyz

```python
# Ä°lk alÄ±ÅŸ ayÄ± (cohort month)
df['CohortMonth'] = (
    df.groupby('Customer ID')['Order Date']
      .transform('min')
      .dt.to_period('M')
)

# SifariÅŸ ayÄ±
df['OrderMonth'] = df['Order Date'].dt.to_period('M')

# Cohort Index (neÃ§É™ ay keÃ§ib)
df['CohortIndex'] = (
    (df['OrderMonth'].dt.year - df['CohortMonth'].dt.year) * 12 +
    (df['OrderMonth'].dt.month - df['CohortMonth'].dt.month) + 1
)

# Cohort data
cohort_data = (
    df.groupby(['CohortMonth', 'CohortIndex'])['Customer ID']
      .nunique()
      .reset_index()
)

# Pivot table (retention matrix)
cohort_counts = cohort_data.pivot(
    index='CohortMonth',
    columns='CohortIndex',
    values='Customer ID'
)

# Cohort size (ilk ay)
cohort_size = cohort_counts.iloc[:, 0]

# Retention rate
retention = cohort_counts.divide(cohort_size, axis=0)

print(retention)


```

```python
plt.figure(figsize=(22, 10))

sns.heatmap(
    retention.iloc[:, :12],   # fewer columns â†’ clearer
    annot=True,
    fmt=".3f",                # more precision
    cmap="viridis",
    vmin=0.15,
    vmax=0.30,
    annot_kws={"size": 9}
)


# sns.heatmap(
#     retention.iloc[:, :12],
#     annot=True,
#     fmt=".3f",
#     cmap="YlGnBu",
#     vmin=0.15,
#     vmax=0.30,
#     annot_kws={
#     "size": 9,
#     "color": "black",
#     "weight": "bold"
# } #It means annotation keywords
# It controls how the numbers inside the heatmap cells look.
# )

plt.title("Cohort Analysis â€“ Customer Retention", fontsize=18)
plt.xlabel("Cohort Index (Months After First Purchase)")
plt.ylabel("Cohort Month (First Purchase)")

plt.show()
```

```python
# Last purchase date per customer
last_purchase = (
    df.groupby('Customer ID')['Order Date']
      .max()
      .reset_index()
)

# Reference date (last date in dataset)
last_date = df['Order Date'].max()

# Inactive days (recency)
last_purchase['inactive_days'] = (
    last_date - last_purchase['Order Date']
).dt.days
```

```python
last_purchase['churned'] = last_purchase['inactive_days'] > 90
churn_rate = last_purchase['churned'].mean() * 100
print(f"Churn Rate: {churn_rate:.1f}%")
```

```python
customer_segment=df[['Customer ID', 'Segment']].drop_duplicates()

churn_seg=last_purchase.merge(customer_segment,on='Customer ID')

churn_by_segment=churn_seg.groupby('Segment')['churned'].mean()*100
print(churn_by_segment)
```

```python
import pandas as pd
df = pd.read_csv('/content/Sample - Superstore.csv',encoding="latin1")  # or your actual file
df['Order Date'] = pd.to_datetime(df['Order Date'])


snapshot_date=df['Order Date'].max() + pd.Timedelta(days=1)

rfm=df.groupby('Customer ID').agg({
    'Order Date' : lambda x:(snapshot_date - x.max()).days, #Recency
    'Order ID' : 'nunique' , #Frequency
    'Sales' : 'sum' #Monetar
    }).reset_index()

rfm.columns=['Customer ID','Recency','Frequency','Monetary']
```

```python
rfm
```

```python
rfm['R_Score']=pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score']=pd.qcut(rfm['Frequency'].rank(method='first'),5,labels=[1,2,3,4,5])
rfm['M_Score']=pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
```

```python
rfm
```

```python
rfm['RFM_Segment']=(
    rfm['R_Score'].astype(str)+
    rfm['F_Score'].astype(str)+
    rfm['M_Score'].astype(str)
)

rfm
```

```python
def segment(row):
  if row['R_Score']==5 and row['F_Score']>=4:
    return 'Champions'
  elif row['R_Score']>=4:
    return 'Loyal'
  elif row['R_Score']<=2  and row['F_Score']>=3:
    return 'At Risk'
  else:
    return 'Others'

rfm['Segment']=rfm.apply(segment, axis=1)
rfm
```

```python
rfm.merge(df[['Customer ID','Sales']],on='Customer ID') \
.groupby('Segment')['Sales'].sum() \
.sort_values(ascending=False)
```

# RFM and Churn Relation

```python
snapshot_date=df['Order Date'].max() + pd.Timedelta(days=1)

last_purchase=df.groupby('Customer ID')['Order Date'].max().reset_index()

last_purchase['days_since_last']=(snapshot_date-last_purchase['Order Date']).dt.days

last_purchase['churned']=last_purchase['days_since_last']>90
```

```python
rfm_churn = rfm.merge(
    last_purchase[['Customer ID','churned']],
    on='Customer ID',
    how='left'
)
rfm_churn
```

```python
churn_table=pd.crosstab(
    rfm_churn['Segment'],
    rfm_churn['churned'],
    normalize='index'

)*100

churn_table
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(churn_table,annot=True,fmt=".1f",cmap="Reds")
plt.title("Churn Rate per Segment (%)")
plt.show()
```

```python
import numpy as np

#Others ve At Risk segmentinden churn olmus musteriler
priority_customers=rfm_churn[
    (rfm_churn['Segment'].isin(['Others','At Risk'])) &
    (rfm_churn['churned']==True)

].copy()

#Monetary +Frequency esasinda prioritetlesdir
priority_customers['Priority_Score']=priority_customers['Monetary']* priority_customers['Frequency']

#priority siralama(en yuksek)
priority_customers=priority_customers.sort_values(by='Priority_Score', ascending=False)

#Yalniz lazim olan sutunlar
priority_list=priority_customers[['Customer ID', 'Segment','Monetary','Frequency','Priority_Score']]
priority_list.head(10) #ilk 10 musteri
```

```python
#CLV (Customer Lifetime Value)--> bir musteri sirket ucun geleckde getireceyi potensial umumi gelir
```

```python
#sade CLV formula: AOV * Frequency * Retention
rfm_churn['AOV']=rfm_churn['Monetary']/rfm_churn['Frequency']

#Retention:churn=True-->0, False--> 1
rfm_churn['Retention']=(~rfm_churn['churned']).astype(int)
rfm_churn['CLV']=rfm_churn['AOV']*rfm_churn['Frequency']*rfm_churn['Retention']

#Segment uzre CLV
segment_clv=rfm_churn.groupby('Segment')['CLV'].mean().sort_values(ascending=False)
segment_clv
```

# A\B test

```python
#others musterileri xilas etmek
```

```python
import numpy as np
others=rfm_churn[rfm_churn['Segment']=='Others'].copy()

np.random.seed(42)
others['group']=np.random.choice(['A','B'],size=len(others))
```

```python
np.random.seed(42)

others['converted']=np.where(
    others['group']=='B',
    np.random.binomial(1, 0.30,len(others)), #B qrupunda 30% conversion,Tekslif alanlar
    np.random.binomial(1, 0.20,len(others)) #A qrupunda 20% conversion
)
```

```python
conversion_rates=others.groupby('group')['converted'].mean()
conversion_rates
```

```python
from statsmodels.stats.proportion import proportions_ztest

succes=others.groupby('group')['converted'].sum()
nobs=others.groupby('group')['converted'].count()

z_stat,p_value=proportions_ztest(succes,nobs)
p_value #0.05<
```
